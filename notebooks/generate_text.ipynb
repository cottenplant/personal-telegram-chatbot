```python
import spacy
from pathlib import Path
import pandas as pd
```

```python
def read_file(filepath):
    with open(input_file_path) as infile:
        str_text = infile.read()
    return str_text


def preprocess(doc_text):
    nlp = spacy.load('en', disable=['parser', 'tagger', 'ner'])
    not_string = '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n '
    return [token.text.lower() for token in nlp(doc_text) if token.text not in not_string]


def create_token_sequence(tokens):
    # because we have such short sentences, should predict less
    train_length = 15 + 1
    text_sequences = []
    for i in range(train_length, len(tokens)):
        seq = tokens[i-train_length:i]
        text_sequences.append(seq)
    return text_sequences

input_file_path = Path('../data/output/samco.txt')
doc_text = read_file(input_file_path)
tokenized_text = preprocess(doc_text)

# will iterate through all text sequences to predict next
token_sequence = create_token_sequence(tokenized_text)
```

```python
pd.DataFrame(token_sequence).to_csv('../seed/text_sequences.csv', index=False)
```
